```python
import pandas as pd
```

##### Creating dataframes


```python
data = {'PoductName': ['Product A', 'Product B', 'Product C', 'Product D'], 'ProductPrice':[2250, 16600, 12500, 5000]}
ProductId = ['A', 'B', 'C', 'D']
df1 = pd.DataFrame(data, index = ProductId)
df1
```

```python
### Creating Data Frames from Pandas Series
ser_products = pd.Series(['Product A', 'Product B', 'Product C', 'Product D'])
ser_prices = pd.Series([22344, 34500, 21000, 45000])

data = {'ProductName': ser_products, 'ProductPrice':ser_prices}
df2 = pd.DataFrame(data)
df2
```

```python
df3 = pd.DataFrame(data = [['Product A', 22250], ['Product B', 16600], ['Product C', 12500]],
                    columns = ['ProductName', 'ProductPrice'],
                    index = ['A', 'B', 'C'])
df3
```
